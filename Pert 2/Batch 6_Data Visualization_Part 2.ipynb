```python
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
plt.figure(figsize=(15,6))
plt.plot(x,cos, linestyle = '-.')
plt.plot(x,sin, linestyle = '--')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Grafik Cos Dan Sin')
plt.show()
```

# Visualisasi Data

```python
df['species'].hist()
plt.show()
```

### scatter plot

```python
# create color dictionary
colors = {'setosa':'r', 'versicolor':'g', 'virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(df['sepal_length'])):
    ax.scatter(df['sepal_length'][i], df['sepal_width'][i],color=colors[df['species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
sns.FacetGrid(df, hue = 'species', size = 6).map(plt.scatter, 'sepal_length','sepal_width' ).add_legend()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['sepal_length'], df['sepal_width'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.show()
```

```python
sns.FacetGrid(df, hue = 'species', size = 6).map(plt.scatter, 'petal_length','petal_width' ).add_legend()
```

```python
df_corr = df.corr()
```

```python
sns.heatmap(data = df_corr, annot = True)
plt.show
```

```python
sns.distplot(df['sepal_length'])
plt.show()
```

```python
sns.distplot(df['sepal_length'], color = 'r', label= 'Sepal Length')
sns.distplot(df['sepal_width'], color = 'b', label= 'Sepal width')
plt.legend()
plt.xlabel('Sepal Width & Sepal Length ')
plt.ylabel('Frekuensi')
plt.show()
```
