```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
plt.figure(figsize=(10, 6))
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Grafik Cos Sin')
plt.plot(x,cos3, marker='',linestyle='--',color='green', label='3 cos x')
plt.plot(x,cos , marker='',linestyle='-', color='red', label = 'cos x')
plt.plot(x,sin , marker='',linestyle='-.', color ='b', label = 'sin x')
plt.plot(x,sin2, marker='x',linestyle='', color= 'yellow', label = '2 sin x')
plt.legend(loc=10)
plt.show()
```

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)

# code here
fig, ax = plt.subplots(nrows=1, ncols=3,sharex=True, sharey=True,figsize=(16,8))
ax[0].plot(x,tan, color='blue',label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x,cos, color='red',label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x,sin, color='green',label='cosX')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

#### Pairplot - Relation, Distribution, Comparison

#### Comparation
Ini digunakan untuk membandingkan besarnya nilai satu sama lain dan untuk dengan mudah mengidentifikasi nilai terendah atau tertinggi dalam data.

Jika Anda ingin membandingkan nilai dari waktu ke waktu, diagram garis atau batang sering kali merupakan pilihan terbaik dan pilihannya bergantung pada jumlah periode yang ingin Anda analisis. Namun, untuk perbandingan antar item, diagram batang atau kolom lebih disukai. Bagan garis memberikan kesan kontinuitas yang mungkin tidak bermakna untuk kategori.


### Komposion
Bagan komposisi digunakan untuk melihat bagaimana sebagian data Anda dibandingkan dengan keseluruhan, dan dapat menunjukkan nilai relatif dan absolut. Mereka dapat digunakan untuk merepresentasikan data statis dan data deret waktu secara akurat.

Untuk data statis, diagram lingkaran bisa melakukan pekerjaan itu. Namun, ada juga opsi lain yang bisa menceritakan kisah yang sama, seperti bagan batang bertumpuk, bagan air terjun, atau peta pohon. Untuk data berbasis waktu, jumlah periode sekali lagi menjadi faktor penentu saat memilih bagan Anda. Anda dapat memvisualisasikan komposisi selama banyak periode dengan bagan area, yang sangat mirip dengan bagan garis, dan bagan batang atau kolom bertumpuk saat Anda memiliki jumlah periode yang berkurang.



### Distribution
Saat mempelajari bagaimana nilai-nilai kuantitatif ditempatkan di sepanjang sumbu, bagan distribusi adalah cara yang harus ditempuh. Dengan melihat bentuk datanya, pengguna dapat mengidentifikasi fitur-fitur seperti rentang nilai, tendensi sentral, dan outlier.

### Relation
Bagan relasi memiliki serangkaian opsi terbatas karena biasanya plot sebar adalah satu-satunya cara yang memadai untuk menyajikan data. Mereka digunakan untuk menemukan korelasi, pencilan, dan cluster dalam data Anda.

Meskipun mata manusia hanya dapat menghargai tiga dimensi secara bersamaan, Anda dapat memvisualisasikan variabel tambahan dengan memetakannya ke ukuran, warna, atau bentuk titik data Anda.

```python
sns.pairplot(data=df_iris, hue = 'class')
plt.show()
```
