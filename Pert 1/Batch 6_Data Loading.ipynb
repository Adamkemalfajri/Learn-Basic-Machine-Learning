```python
mean_age = df['Age'].mean()
mean_age
```

```python
df['Age'] = df['Age'].fillna(mean_age)
df.isnull().sum()
```

```python
df = df.drop('Cabin', axis = 1)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].astype('int64')
```

```python
df.info()
```

```python
df['Pclass'].unique()
```

```python
df['Pclass'].value_counts()
```

```python
df['Sex'].unique()
```

```python
df['Sex'].value_counts()
```

```python
new_columns = {
 'PassengerId': 'PASS_ID',
 'Survived': 'SURVIVED', 
 'Pclass': 'PCLASS', 
 'Name': 'NAME', 
 'Sex': 'SEX', 
 'Age': 'AGE', 
 'SibSp': 'SIBSA', 
 'Parch': 'PARCA' , 
 'Fare': 'FARE',
 'Ticket' : 'TICKET',
 'Embarked' : "EMBARKED"
}
```

```python
df = df.rename(columns = new_columns)
df.head()
```

```python
df = df.rename(columns = {'PCLASS':'Passenger Class', 'SEX' : 'Jenis Kelamin'})
df.head(1)
```

```python
df['Jenis Kelamin'] = df['Jenis Kelamin'].replace(['female','male'],['Woman','Man'])
df.head(2)
```

```python
part1 = df[["SURVIVED", "Passenger Class", "AGE","FARE"]]
sns.pairplot(part1,size=4)
plt.show()
```
