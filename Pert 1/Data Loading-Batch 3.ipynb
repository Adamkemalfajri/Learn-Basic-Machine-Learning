## Rename Columns

```python
new_columns = {
 'PassengerId': 'PASS_ID',
 'Survived': 'SURVIVED', 
 'Pclass': 'PCLASS', 
 'Name': 'NAME', 
 'Sex': 'SEX', 
 'Age': 'AGE', 
 'SibSp': 'SIBSA', 
 'Parch': 'PARCA' , 
 'Fare': 'FARE',
 'Ticket' : 'TICKET',
 'Embarked' : "EMBARKED"
}
```

```python
df = df.rename(columns = new_columns )
df.head()
```

```python
df = df.set_index('PASS_ID')
df.head()
```

## PIVOT

```python
data = {
    'foo' : ['one']*3 + ['two']*3,
    'bar' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'baz' : [1, 2, 3, 4, 5, 6],
    'zoo' : ['x', 'y', 'z', 'q', 'w', 't']
}
df_new = pd.DataFrame(data)
df_new
```

```python
pivot_df = pd.pivot(df_new, index = "foo", columns="bar", values='baz')
pivot_df
```
